# Naive Bayes


```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import *
import warnings


warnings.filterwarnings('ignore')
```

```python
col_names = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']
```

```python
df = pd.read_csv("data/car.csv",names = col_names)
```

# EDA

```python
df.describe().T.style.background_gradient(cmap='Greens')
```

```python
def plot_bar(df,i):
    #fig = plt.figure(figsize = (7, 6))
    plt.subplot(1,1,1)
    sns.countplot(df[i])
    plt.title(f"Distribution of {i}")
    plt.show()
for i in df.columns:
    plot_bar(df,i)
```

```python
li = list(df.columns)
li
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
for i in li:
    df[i] = labelencoder.fit_transform(df[i])
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30 ,random_state =1)
```

# Naive Bayes

```python
from sklearn.naive_bayes import *
nb=GaussianNB()
nb.fit(x_train,y_train)
y_pred4 = nb.predict(x_test)
```

```python
plt.title('Naive Bayes Confusion Matrix')
sns.heatmap(confusion_matrix(y_test,y_pred4),annot=True,fmt='g')
plt.show()
```

```python
print(f"Naive Bayes accuracy is : {round(accuracy_score(y_test,y_pred4)*100,2)} %")
```

```python
print(classification_report(y_test,y_pred4))
```
